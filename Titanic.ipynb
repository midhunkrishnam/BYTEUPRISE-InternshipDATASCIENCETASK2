```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/tested.csv")
df
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Fare'].fillna(df['Fare'].median(), inplace=True)
df.dropna(subset=['Cabin'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True) # Convert categorical variables to numerical using one-hot encoding
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Pclass', y='Survived', data=df, color='red', ci=None)
plt.title('Survival Rate by Passenger Class', color='green')
plt.show()
```

```python

sns.histplot(df['Age'], bins=20, kde=True,color='blue')
plt.title('Age Distribution of Passengers',color='green')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python

sns.histplot(df['Fare'], bins=20, kde=True,color='blue')
plt.title('Fare Distribution',color='green')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
sns.barplot(x='FamilySize', y='Survived', data=df, color='red', ci=None)
plt.title('Survival Rate by Family Size', color='green')
plt.xlabel('Family Size')
plt.ylabel('Survival Rate')
plt.show()
```

```python

survived_count = df['Survived'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(survived_count, labels=['Not Survived', 'Survived'], autopct='%1.1f%%', colors=['red', 'lightgreen'], startangle=140)
plt.title('Survival Rate',color='green')
plt.axis('equal')
plt.show()
```

```python

plt.figure(figsize=(8, 6))
plt.scatter(df['Age'],df['Fare'], alpha=0.5, color='red')
plt.title('Scatter Plot of Age vs Fare',color='green')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']])
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Age', data=df,color='red')
plt.title('Box Plot of Age by Passenger Class',color='green')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
correlation_matrix = df.corr()

#HeatMap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap',color="green")
plt.show()
```
